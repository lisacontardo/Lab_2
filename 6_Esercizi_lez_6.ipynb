### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

#Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
k = 1
np.random.seed(42) 
campione = gamma.rvs(a=k, size=1000)
```

```python
#stimo il parametro di forma del campione usando metodo fit
shape_fit, loc_fit, scale_fit = gamma.fit(campione, floc=0)
print(f"Parametro di forma stimato: {shape_fit:}")
```

```python
#Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.
x = np.linspace(0, campione.max(), 1000)
pdf = gamma.pdf(x, a=shape_fit, scale=scale_fit)

plt.figure(figsize=(8,5))
plt.hist(campione, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Campione')
plt.plot(x, pdf, 'r', lw=2, label='PDF stimata')
plt.title('Istogramma e PDF della distribuzione gamma')
plt.xlabel('Valore')
plt.ylabel('Densità')
plt.legend()
plt.show()
```

```python
#Traccia la funzione di distribuzione cumulativa (CDF).

x = np.linspace(0, campione.max(), 1000)
cdf = gamma.cdf(x, a=shape_fit, scale=scale_fit)

plt.figure(figsize=(8,5))
plt.plot(x, cdf, 'g', lw=2, label='CDF stimata')
plt.title('Funzione di distribuzione cumulativa')
plt.xlabel('Valore')
plt.ylabel('Probabilità cumulativa')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Calcola la varianza.
varianza = np.var(campione, ddof=1)  # ddof=1 per varianza campionaria
print(f"Varianza del campione: {varianza}")
```

### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])

import matplotlib.pyplot as plt

months = np.arange(12)
plt.plot(months, temp_max, "ro") #ro = punti rossi
plt.plot(months, temp_min, "bo") #bu = punti blu
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
```

```python
# dominio 
mesi_interpolati = np.linspace(0, 11, 100) 
# calcola i coefficienti del polinomio
coefficienti_massime = np.polyfit(months, temp_max, 4)  
#valuto il polinomio nei 100 punti 
curva_massime = np.polyval(coefficienti_massime, mesi_interpolati) 

#stessa cosa per le temperature minime
coefficienti_minime = np.polyfit(months, temp_min, 4)
curva_minime = np.polyval(coefficienti_minime, mesi_interpolati)

#ridisegno il grafico
plt.title("Interpolazione numpy")
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")

#Disegno le curve interpolate
plt.plot(mesi_interpolati, curva_massime, color = 'orange')
plt.plot(mesi_interpolati, curva_minime, color = 'cyan')
plt.show()

#valuto il polinomio ottenendo le stime cioè i valori predetti
stime_temp_max = np.polyval(coefficienti_massime, months)
stime_temp_min = np.polyval(coefficienti_minime, months)
```

```python
#calcolo MAE e RMSE

mae_massime = np.mean(np.abs(temp_max - stime_temp_max))
rmse_massime = np.sqrt(np.mean((temp_max - stime_temp_max) ** 2))

mae_minime = np.mean(np.abs(temp_min - stime_temp_min))
rmse_minime = np.sqrt(np.mean((temp_min - stime_temp_min) ** 2))

print("MAE temperature massime: ", mae_massime)
print("RMSE temperature massime: ", rmse_massime)
print("MAE temperature minime: ", mae_minime)
print("RMSE temperature minime: ", rmse_minime)
```

```python
#fit sinusoidale
from scipy.optimize import curve_fit

def funzione_seno(x, A, B, C, D):
    return A * np.sin(B * x + C) + D

parametri_max, _ = curve_fit(funzione_seno, months, temp_max, p0=[10, 2*np.pi/12, 0, 20])
parametri_min, _ = curve_fit(funzione_seno, months, temp_min, p0=[30, 2*np.pi/12, 0, -30])

x_fit = np.linspace(0, 11, 300)
fit_max = funzione_seno(x_fit, *parametri_max)
fit_min = funzione_seno(x_fit, *parametri_min)

plt.figure(figsize=(10, 6))
plt.plot(months, temp_max, "ro", label="Temp max osservate")
plt.plot(months, temp_min, "bo", label="Temp min osservate")
plt.plot(x_fit, fit_max, color="orange", label="Fit sinusoidale max")
plt.plot(x_fit, fit_min, color="cyan", label="Fit sinusoidale min")
plt.xlabel("Mese")
plt.ylabel("Temperatura")
plt.title("Fit sinusoidale delle temperature mensili")
plt.legend()
plt.grid(True)
plt.show()

# Stime nei punti originali
stime_max = funzione_seno(months, *parametri_max)
stime_min = funzione_seno(months, *parametri_min)
```

```python
#valutare l'efficenza della curva
#funzione per calcolare MAE e RMSE
mae_max = np.mean(np.abs(temp_max - stime_max))
rmse_max = np.sqrt(np.mean((temp_max - stime_max) ** 2))
mae_min = np.mean(np.abs(temp_min - stime_min))
rmse_min = np.sqrt(np.mean((temp_min - stime_min) ** 2))

print("MAE temp max:", mae_max)
print("RMSE temp max:", rmse_max)
print("MAE temp min:", mae_min)
print("RMSE temp min:", rmse_min)
```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
#regressione 
import numpy as np 
from scipy.stats import linregress
#variabile indipendente
x = df['mpg'] 
y = df['disp']
slope, intercept, r_value, p_value, std_err = linregress(x, y)

print("Slope: ", slope)
print("Intercept", intercept)
print("R-quadro:", r_value**2)
print("p-value:", p_value)

#predizione dei valori stimati (equazione della retta)
y_pred = slope * x + intercept

plt.scatter(x, y, label='Dati reali')
plt.plot(x, y_pred, color='red', label='Regressione lineare')
plt.xlabel("Consumo (mpg)")
plt.ylabel("Cilindrata (disp)")
plt.title("Modello di regressione lineare: disp ~ mpg")
plt.legend()
plt.grid(True)
plt.show()
```
