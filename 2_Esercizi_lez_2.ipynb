### **1. Indicizzazione booleana**

1. Creare un nuovo vettore con tutti i numeri primi tra 0 e 20. I numeri primi compresi tra 0 e 20 sono: **2, 3, 5, 7, 11, 13, 17, 19**.

2. Utilizza un test logico (ad esempio `>`, `<`, `==`, `!=`) per **selezionare tutte le voci del vettore maggiori di 10**.

3. Utilizza un test logico per selezionare **tutti i numeri primi pari** dalla lista.  


```python
import numpy as np
#1
vettore = np.array([2, 3, 5, 7, 11, 13, 17, 19])
#2
a = vettore[vettore > 10]
print(a)
#3
b = vettore[vettore % 2 == 0]
print(b)
```

### **3. Seleziona il numero più vicino a 0.5 in un array casuale**
Genera un array 10 x 3 con numeri casuali nell'intervallo [0,1].
Per ogni riga, seleziona il numero più vicino a 0.5. Per farlo usa il fancy indexing.

```python
v = np.random.rand(10, 3) #genera di default numeri tra 0 e 1
print(v)

colonne = np.abs(v - 0.5).argmin(axis=1) #prendi l’indice della colonna col valore più piccolo per ogni riga
righe = np.arange(10)  

v1 = v[righe, colonne]
print(v1)
```

1. Creare un vettore contenente i seguenti stipendi: **50.000, 105.250, 55.000, 89.000**.  
Qual è il costo totale del personale (somma di tutti gli stipendi dell'azienda)?

2. Supponiamo che il nostro malvagio CEO abbia deciso di darsi un aumento.  
Modifica il vettore degli stipendi in modo che il CEO, la persona che guadagna **105.250 dollari**, riceva un aumento del **15%**.

3. Il **115%** di **105.250 dollari** è **121.037,50 dollari**.

4. Ricrea il vettore e utilizza l'argomento **`dtype`** in modo che, quando il CEO riceve un aumento del **15%**, il suo stipendio finale sia esattamente **121.037,50 dollari**.

5. Questo aumento ha infastidito l'impiegata con lo stipendio più basso (**50.000 dollari**), che ora pretende un aumento del **20%**.  
Modifica il vettore per riflettere questo cambiamento.

6. L'aumento dato al CEO e all'impiegata ha irritato gli altri due dipendenti, quindi ora devono ricevere un aumento del **10%**.  
Modifica il vettore di conseguenza.

7. Calcola il **costo totale del personale** dopo tutti gli aumenti.  
Alla fine, quanto è costato all'azienda l'aumento di circa **16.000 dollari** del CEO?


```python
import numpy as np

#1 Creazione del vettore degli stipendi (uso float per evitare troncamenti)
vettore = np.array([50000, 105250, 55000, 89000])
print("Stipendi:", vettore)
# Costo totale iniziale
somma = np.sum(vettore)
print("Costo totale del personale:", somma)

#2 Aumento del 15% al CEO (105.250 $)
#vettore[vettore == 105250] *= 1.15
#np.where(condizione, valore_se_vero, valore_se_falso)
vettore2 = np.where(vettore == 105250, 105250 * 1.15, vettore)
print("stipendi dopo l'aumento del CEO:", vettore2)

#3
print(vettore2 == 121037.50)
#non risulta presente ed è dovuto alla precisione dei numeri floating point
#si potrebbe risolvere usando nd.round(105250 * 1.15, 2)


#modifico solo l'array stipendi senza crearn di nuovi
#4 
stipendi = np.array([50000, 105250, 55000, 89000], dtype=np.float64)
stipendi = np.where(stipendi == 105250, 105250 * 1.15, stipendi)
print("Stipendi dopo l'aumento del CEO:", stipendi)
#controllo se mi ritorna True
print(stipendi == 121037.50)
#non ritorna true perchè non vengono rappresentati tutti i numeri decimali con precisione 
#quindi utilizziamo isclose
print(np.isclose(stipendi, 121037.50))

#5 Aumento del 20% alla dipendente con lo stipendio più basso (50.000 $)
stipendi[stipendi == 50000] *= 1.20

#6 Aumento del 10% agli altri due dipendenti
stipendi[(stipendi == 55000) | (stipendi == 89000)] *= 1.10
print("Stipendi dopo gli aumenti a tutti i dipendenti:\n", stipendi)

#7 Costo totale finale
costo_finale = stipendi.sum()
print("Costo totale finale del personale:\n", costo_finale)

# Quanto è costato all'azienda l'aumento di 16000 dollari dell'aumento del CEO?
aumento_ceo = 121037.5 - 105250
print("Costo dell'aumento del CEO:", aumento_ceo)

```

Data la matrice survey con età, reddito e anni di istruzione

    [25, 40000, 10],
    [32, 52000, 12],
    [40, 63000, 14],
    [29, 47000, 11],
    [35, 58000, 13]
])


1. Seleziona dalla matrice solo gli intervistati con **almeno 12 anni di istruzione** utilizzando un test logico.

2. In un'unica riga di codice, seleziona i **redditi** degli intervistati con **almeno 12 anni di istruzione**.

3. In un'unica riga di codice, calcola il **reddito medio** degli intervistati con **almeno 12 anni di istruzione**.


```python
import numpy as np

survey_matrix = np.array([
    [25, 40000, 10],
    [32, 52000, 12],
    [40, 63000, 14],
    [29, 47000, 11],
    [35, 58000, 13]
    ])

#1
istruzione = survey_matrix[survey_matrix[:, 2] >= 12]
print("Intervistati con istruzione ≥ 12:")
print(istruzione)
#2 
#condizione sulle righe e seleziono la colonna dei redditi
redditi = survey_matrix[survey_matrix[:, 2] >= 12, 1]
#survey_matrix[:, 2] >= 12 è un array di True e False
print("Redditi:")
print(redditi)

#3
reddito_medio = survey_matrix[survey_matrix[:, 2] >= 12, 1].mean()
print("Reddito medio:")
print(reddito_medio)
```
